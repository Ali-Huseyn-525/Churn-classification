# CHANGE data-type

```python
data['Ümumi ödənişlər'] = pd.to_numeric(data['Ümumi ödənişlər'], errors='coerce')
```

# NULL ımpute

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='most_frequent')
data['cinsiyyət'] = imputer.fit_transform(data[['cinsiyyət']]).ravel()
data['tərəfdaş'] = imputer.fit_transform(data[['tərəfdaş']]).ravel()
```

```python
data = data.dropna(subset=['Ümumi ödənişlər'])
```

```python
data.isnull().sum()
```

# CHECK OUTLIER


```python
import matplotlib.pyplot as plt
plt.boxplot(data['müddəti'])
plt.show()
```

```python
Q1 = data['Aylıq Ödənişlər'].quantile(0.25)
Q3 = data['Aylıq Ödənişlər'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers1 = data[(data['Aylıq Ödənişlər'] < lower_bound) | (data['Aylıq Ödənişlər'] > upper_bound)]

print(outliers1)
```

```python
from scipy.stats import zscore
outliers2 = data[abs(zscore(data['Ümumi ödənişlər'])) > 3]
print(outliers2)
```

# ENCODING

```python
from sklearn.preprocessing import LabelEncoder

categorical_columns = data.select_dtypes(include=['object']).columns
binary_columns = categorical_columns[(data[categorical_columns].nunique() == 2) & (categorical_columns != 'Churn')]

label_encoders = {}

for column in binary_columns:
  le = LabelEncoder()
  data[column] = le.fit_transform(data[column])
  label_encoders[column] = le
```

```python
import pandas as pd

categorical_columns = data.select_dtypes(include=['object']).columns
non_binary_columns = categorical_columns[data[categorical_columns].nunique() > 2]

data = pd.get_dummies(data, columns=non_binary_columns, drop_first=False)
```

```python
data['Churn'] = data['Churn'].replace({'Hə': 1, 'Yox': 0})
```

```python
data.info()
```

# FEATURE SCALING

```python
from sklearn.preprocessing import StandardScaler

columns_to_normalize = ['müddəti', 'Aylıq Ödənişlər','Ümumi ödənişlər']

scaler = StandardScaler()
data[columns_to_normalize] = scaler.fit_transform(data[columns_to_normalize])
```

```python
data.head()
```

```python
pd.DataFrame(data).to_excel("dataset_preprocessing_N1.xlsx",index=False)
```

# MODEL BUILDING

```python
from sklearn.model_selection import train_test_split
X = data.drop('Churn', axis=1)
y = data['Churn']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

# MODEL OPTIMIZATION

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=42)

X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)
```

```python
y_train_resampled.value_counts()
```

#  MODEL BUILDING_RandomForest

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
from sklearn.ensemble import RandomForestClassifier

rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(X_train_resampled, y_train_resampled)


y_pred_rf = rf_model.predict(X_train_resampled)
y_pred_rf_prob = rf_model.predict_proba(X_train_resampled)[:, 1]
rf_accuracy = accuracy_score(y_train_resampled, y_pred_rf)
rf_precision = precision_score(y_train_resampled, y_pred_rf)
rf_recall = recall_score(y_train_resampled, y_pred_rf)
rf_f1 = f1_score(y_train_resampled, y_pred_rf)
rf_roc_auc = roc_auc_score(y_train_resampled, y_pred_rf_prob)

print("Random Forest Model Metrics:")
print("Accuracy:", rf_accuracy)
print("Precision:", rf_precision)
print("Recall:", rf_recall)
print("F1 Score:", rf_f1)
print("ROC AUC Score:", rf_roc_auc)
```

# MODEL BUILDING_SVC

```python
from sklearn.svm import SVC

svm_model = SVC(probability=True, random_state=42)
svm_model.fit(X_train_resampled, y_train_resampled)


y_pred_svm = svm_model.predict(X_train_resampled)
y_pred_svm_prob = svm_model.predict_proba(X_train_resampled)[:, 1]
svm_accuracy = accuracy_score(y_train_resampled, y_pred_svm)
svm_precision = precision_score(y_train_resampled, y_pred_svm)
svm_recall = recall_score(y_train_resampled, y_pred_svm)
svm_f1 = f1_score(y_train_resampled, y_pred_svm)
svm_roc_auc = roc_auc_score(y_train_resampled, y_pred_svm_prob)

print("SVM Model Metrics:")
print("Accuracy:", svm_accuracy)
print("Precision:", svm_precision)
print("Recall:", svm_recall)
print("F1 Score:", svm_f1)
print("ROC AUC Score:", svm_roc_auc)
```

# MODEL BUILDING_xgboost

```python
from xgboost import XGBClassifier
xgb_model = XGBClassifier(random_state=42, use_label_encoder=False, eval_metric='logloss')
xgb_model.fit(X_train_resampled, y_train_resampled)

y_pred_xgb = xgb_model.predict(X_train_resampled)
y_pred_xgb_prob = xgb_model.predict_proba(X_train_resampled)[:, 1]
xgb_accuracy = accuracy_score(y_train_resampled, y_pred_xgb)
xgb_precision = precision_score(y_train_resampled, y_pred_xgb)
xgb_recall = recall_score(y_train_resampled, y_pred_xgb)
xgb_f1 = f1_score(y_train_resampled, y_pred_xgb)
xgb_roc_auc = roc_auc_score(y_train_resampled, y_pred_xgb_prob)

print("XGBoost Model Metrics:")
print("Accuracy:", xgb_accuracy)
print("Precision:", xgb_precision)
print("Recall:", xgb_recall)
print("F1 Score:", xgb_f1)
```

# MODEL BUILDING_LogisticRegression

```python
from sklearn.linear_model import LogisticRegression

lr_model = LogisticRegression(random_state=42, max_iter=1000)
lr_model.fit(X_train_resampled, y_train_resampled)

y_pred_lr = lr_model.predict(X_train_resampled)
y_pred_lr_prob = lr_model.predict_proba(X_train_resampled)[:, 1]
lr_accuracy = accuracy_score(y_train_resampled, y_pred_lr)
lr_precision = precision_score(y_train_resampled, y_pred_lr)
lr_recall = recall_score(y_train_resampled, y_pred_lr)
lr_f1 = f1_score(y_train_resampled, y_pred_lr)
lr_roc_auc = roc_auc_score(y_train_resampled, y_pred_lr_prob)

print("Logistic Regression Model Metrics:")
print("Accuracy:", lr_accuracy)
print("Precision:", lr_precision)
print("Recall:", lr_recall)
print("F1 Score:", lr_f1)
```

```python
from sklearn.model_selection import train_test_split
from lazypredict.Supervised import LazyClassifier
from imblearn.over_sampling import SMOTE
import pandas as pd

assert 'Churn' in data.columns, "Column 'Churn' not found in dataset!"

X = data.drop(columns=['Churn'])
y = data['Churn']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

smote = SMOTE(random_state=42)
X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)

lazy_clf = LazyClassifier(verbose=0, ignore_warnings=True, custom_metric=None)

models, predictions = lazy_clf.fit(X_train_resampled, X_test, y_train_resampled, y_test)

print("Lazy Prediction After SMOTE:")
print(models)

models.to_csv('lazy_prediction_results.csv', index=False)
print("Results saved to 'lazy_prediction_results.csv'")
```
